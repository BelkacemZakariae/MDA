```python
import paramiko
from datetime import datetime
import os
import logging
import pandas as pd

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

SFTP_HOST_DOWNLOAD = 'sftp-partners.glovoapp.com'
SFTP_PORT_DOWNLOAD = 22
SFTP_USERNAME_DOWNLOAD = 'carrefour-ma'
SFTP_PASSWORD_DOWNLOAD = 'l7ArAHotIzDu6U2uYisRGrUt'

download_path = r"C:\Users\Zakar\OneDrive\Desktop\MDA"
date_today = datetime.now().strftime('%d%m%y')  
file_prefix = "R_carrefour_stock_"
remote_directory = "/glovoapp-partners-sftp-bucket-121e9009/home/carrefour-ma/input_carrefour-ma"

try:
    transport_download = paramiko.Transport((SFTP_HOST_DOWNLOAD, SFTP_PORT_DOWNLOAD))
    transport_download.connect(username=SFTP_USERNAME_DOWNLOAD, password=SFTP_PASSWORD_DOWNLOAD)
    sftp_download = paramiko.SFTPClient.from_transport(transport_download)

    logger.info(f"Connected to SFTP server for downloading: {SFTP_HOST_DOWNLOAD}")

    sftp_download.chdir(remote_directory)

    remote_files = sftp_download.listdir()
    logger.info(f"Files on server in {remote_directory}: {', '.join(remote_files)}")  

   
    downloaded_files = []
    count = 0

    for remote_file in remote_files:
        if count >= 6: 
            break

        logger.info(f"Checking file: {remote_file}") 

        if remote_file.startswith(file_prefix) and date_today in remote_file and remote_file.endswith('.csv'):
            file_info = sftp_download.stat(remote_file)
            upload_time = datetime.fromtimestamp(file_info.st_mtime)  


            if upload_time.hour > 6:
                local_file_path = os.path.join(download_path, remote_file)
                sftp_download.get(remote_file, local_file_path)
                downloaded_files.append(local_file_path)
                logger.info(f"Downloaded file: {remote_file} to {local_file_path}")
            else:
                logger.warning(f"File {remote_file} was uploaded after 6 AM.")

            count += 1 

    sftp_download.close()
    transport_download.close()

except paramiko.AuthenticationException:
    logger.error("Authentication failed. Please check your username and password for download.")
except Exception as e:
    logger.error(f"An error occurred during the download process: {str(e)}")

if downloaded_files:
    store_ids = [195, 166, 106, 746, 177, 741, 744, 187, 657, 652, 135]
    combined_filtered_df = pd.DataFrame()

    for file in downloaded_files:
        df = pd.read_csv(file)
        
        for store_id in store_ids:
            filtered_df = df[df['store_id'] == store_id]
            if not filtered_df.empty:
                combined_filtered_df = pd.concat([combined_filtered_df, filtered_df], ignore_index=True)
                logger.info(f"Filtered data for store_id {store_id} from file: {file}")

    if not combined_filtered_df.empty:
        consolidated_file_path = os.path.join(download_path, f"carrefour_stock_{date_today}.xlsx")
        combined_filtered_df.to_excel(consolidated_file_path, index=False)
        logger.info(f"Consolidated filtered data saved to: {consolidated_file_path}")
    else:
        logger.warning("No filtered data found for the specified store_ids.")

    SFTP_HOST_UPLOAD = 'ftp-carrefour.ora.ma'
    SFTP_PORT_UPLOAD = 22
    SFTP_USERNAME_UPLOAD = 'ora-ftp'
    SFTP_PASSWORD_UPLOAD = 'gEEw9KeK5^9D^ZWK'
    
    upload_path = f'/public_ftp/carrefour_stock_{date_today}.xlsx'

    try:
        transport_upload = paramiko.Transport((SFTP_HOST_UPLOAD, SFTP_PORT_UPLOAD))
        transport_upload.connect(username=SFTP_USERNAME_UPLOAD, password=SFTP_PASSWORD_UPLOAD)
        sftp_upload = paramiko.SFTPClient.from_transport(transport_upload)
        
        logger.info(f"Connected to SFTP server for uploading: {SFTP_HOST_UPLOAD}")

        sftp_upload.put(consolidated_file_path, upload_path)
        logger.info(f"Uploaded file: carrefour_stock_{date_today}.xlsx to {SFTP_HOST_UPLOAD}{upload_path}")

        sftp_upload.close()
        transport_upload.close()

    except paramiko.AuthenticationException:
        logger.error("Authentication failed. Please check your username and password for upload.")
    except Exception as e:
        logger.error(f"An error occurred during the upload process: {str(e)}")

```
